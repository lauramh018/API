```python
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import time
import key
from citipy import citipy

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
base_url= "https://api.openweathermap.org/data/2.5/weather?"
units="Metric"
```

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
#Avoiding the space error
    city = city.replace(" ", "%20")
    
# If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
print(cities)
```

```python
#My key
api_key=key.key
```

```python
#Lists for our variables
city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

# Start the call counter 
loop = 1

# Log file print statement
print(f"Trying to retrieve data")
print(f"-------------------------------")


for city in cities:
    try: 
        query_url=f'{base_url}appid={api_key}&q={city}&units={units}'
        response=requests.get(query_url).json()
        city_name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        city_record = response["name"]
        print(f"Record number: {loop} | {city_record}")
        print(f"{query_url}&q={city}")
        
        #Pass to the next record
        loop= loop + 1
        
    except:
        print("Upps...not found. Next...")
    continue
```

```python
#Create a dictionary
weather_dict ={
    "City": city_name,
    "Cloudiness":cloudiness, 
    "Country":country,
    "Date":date, 
    "Humidity": humidity,
    "Lat":lat, 
    "Lng":lng, 
    "Max Temp": max_temp,
    "Wind Speed":wind_speed
}

#Cnvert to Dataframe
weather_data = pd.DataFrame(weather_dict)

weather_data.head(10)
```
